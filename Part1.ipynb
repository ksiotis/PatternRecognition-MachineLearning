## Ερώτημα 2: Υπολογισμός παραγώγων

### 2.1 Έστω $f(x)=x^TAx+b^Tx$, μια συνάρτηση d μεταβλητών, με $A$ συμμετρικό πίνακα (δηλ. $A=A^T$) διάστασης d x d, και $b\in\mathbb{R}^d$. Να υπολογίσετε αναλυτικά βήμα προς βήμα με το χέρι, την παράγωγο της $f(x)$ ως προς το διάνυσμα $\in\mathbb{R}^d$ παραθέτωντας όλα τα ενδιάμεσα βήματα.

$$\nabla f(x)=\nabla(x^TAx+b^Tx)=\nabla(x^TAx)+\nabla(b^Tx)\qquad\qquad(1)$$ 
  
$$\nabla(b^Tx)=\nabla(b_1x_1+\cdots b_dx_d)=
\begin{pmatrix}
\frac{\vartheta}{\vartheta x_1}(b_1x_1+\cdots+b_dx_d)\\
\vdots\\
\frac{\vartheta}{\vartheta x_d}(b_1x_1+\cdots+b_dx_d)
\end{pmatrix}=
\begin{pmatrix}
b_1\\
\vdots\\
b_d
\end{pmatrix}
=b\qquad\qquad(2)$$  
$$\nabla(x^TAx)=\nabla x^T
\begin{pmatrix}
x_1A_{11}+\cdots+x_1A_{1d}\\
\vdots\\
x_dA_{d1}+\cdots+x_dA_{dd}
\end{pmatrix}=
\nabla x_1(x_1A_{11}+\cdots+x_dA_{1d})+\cdots+x_d(x_1A_{d1}+\cdots+x_dA_{dd})=
\begin{pmatrix}
\frac{\vartheta}{\vartheta x_1}x_1(x_1A_{11}+\cdots+x_dA_{1d})+\cdots+x_d(x_1A_{d1}+\cdots+x_dA_{dd})\\
\vdots\\
\frac{\vartheta}{\vartheta x_d}x_1(x_1A_{11}+\cdots+x_dA_{1d})+\cdots+x_d(x_1A_{d1}+\cdots+x_dA_{dd})
\end{pmatrix}=
\begin{pmatrix}
2x_1A_{11}+\cdots+x_kA_{1k}+x_kA_{k1}+\cdots+x_dA_{1d}+x_dA_{d1}\\
\vdots\\
x_1A_{k1}+x_1A_{1k}+\cdots+2x_kA_{kk}+\cdots+x_dA_{kd}+x_dA_{dk}\\
\vdots\\
x_1A_{d1}+x_1A_{1d}+\cdots+x_kA_{dk}+x_kA_{kd}+\cdots+2x_dA_{dd}
\end{pmatrix}=
\begin{pmatrix}
x_1(A_{11}+A_{11})+\cdots+x_k(A_{1k}+A_{k1})+\cdots+x_d(A_{1d}+A_{d1})\\
\vdots\\
x_1(A_{k1}+A_{1k})+\cdots+x_k(A_{kk}+A_{kk})+\cdots+x_d(A_{kd}+A_{dk})\\
\vdots\\
x_1(A_{d1}+A_{1d})+\cdots+x_k(A_{dk}+A_{kd})+\cdots+x_d(A_{dd}+A_{dd})
\end{pmatrix}=(A+A^T)x\qquad\qquad(3)$$  
Άρα:  
$$(1)\stackrel{(2),(3)}{\Longrightarrow}\nabla f(x)=(A+A^T)x+b$$
Όμως $A$ συμμετρικός, άρα $Α=Α^Τ$  
Οπότε: $$\nabla f(x)=2Ax+b$$

### 2.2 Έστω $Α, Β, Χ$ πίνακες πραγματικών αριθμών διάστασης d x d με τάξη (rank) ίση με d. Nα βρείτε, αναλυτικά και σε κλειστή μορφή (closed form solution) το ολικό ελάχιστο του παρακάτω προβλήματος βελτιστοποίησης:$$\min_X\|A-XB\|_F^2$$Να εξηγήσετε με λεπτομέρεια όλη τη διαδικασία σκέψης που θα ακολουθήσετε και να παραθέσετε όλα τα ενδιάμεσα βήματα των πράξεων.

$$\|A-XB\|_F^2=\sqrt{\sum_{i=1}^d\sum_{j=1}^d[A_{ij}-(XB)_{ij}]^2}^2=\sum_{i=1}^d\sum_{j=1}^d[A_{ij}-(XB)_{ij}]^2\qquad\qquad(1)$$
Ως άθροισμα τετραγώνων η ελάχιστη τιμή της $(1)$ θα είναι 0 αν κάθε επιμέρους όρος είναι 0. Συνεπώς πρέπει:  
$$\min_X\|A-XB\|_F^2\Rightarrow A-XB=0\Rightarrow A=XB\qquad\qquad(2)$$
Αφού ο πίνακας $B$ είναι d x d  και $rank=d$ σημαίνει ότι η ορίζουσά του είναι $det\neq0$, άρα υπάρχει ο αντίστροφος πίνακας $B^{-1}$. Με βάση αυτό, έχουμε:
$$(2)\stackrel{\exists B^{-1}}{\Longrightarrow}AB^{-1}=XBB^{-1}=XI_d=Χ\Rightarrow X=AB^{-1}$$

## Ερώτημα 3: Gradient descent

```python
def f1(x1, x2):
        return (x1-2)**2+(x2-3)**2

def df1_1(x1, x2):
    return 2.0*x1-4

def df1_2(x1, x2):
    return 2.0*x2-6

def f2(x1, x2):
    return (1.0-(x2-3.0))**2+20.0*((x1+3.0)-(x2-3.0)**2)**2

def df2_1(x1, x2):
    return 40.0*x1+240.0*x2-40.0*x2**2-240.0

def df2_2(x1, x2):
    return 240.0*x1-80.0*x1*x2+80.0*x2**3+1922.0*x2-720.0*x2**2-1448.0

class gradient_descent:
    def __init__(self, func, learningRate, maxIts = 10):
        if func == 'f1':
            self.f = f1
            self.df1 = df1_1
            self.df2 = df1_2
        elif func == 'f2':
            self.f = f2
            self.df1 = df2_1
            self.df2 = df2_2
        self.w = [0.0, 0.0]
        self.lr = learningRate
        self.T = maxIts

    def weightChange(self, x1, x2):
        temp = [-self.lr*self.df1(x1, x2), -self.lr*self.df2(x1, x2)]
        # print(x1,x2,temp)
        return temp

    def iteration(self):
        try:
            deltas = self.weightChange(self.w[0], self.w[1])
            self.w[0] = self.w[0] + deltas[0]
            self.w[1] = self.w[1] + deltas[1]
            return self.f(self.w[0], self.w[1])
        except OverflowError:
            # print('Overflow at (',self.w[0],',',self.w[1],')')
            return None
        

    def descent(self):
        res = [self.f(0,0)]
        for i in range(self.T):
            t = self.iteration()
            res.append(t)
        return res

    def plotDescent(self):
        d = self.descent()
        plt.plot(d)
        plt.title('learning rate '+str(self.lr))
        if (d[-1] == None):
            plt.yscale('log')
        plt.show()
```

#### f1(x) for learning rate 0.5

```python
gradient_descent('f1', 0.5).plotDescent()
```

#### f2(x) for learning rate 0.5

```python
gradient_descent('f2', 0.5).plotDescent()
```

#### f1(x) for various learning rates

```python
for lr in [1.5, 1.0, 0.5, 0.1, 0.05, 0.01, 0.005, 0.001, 0.0005, 0.0001]:
    gradient_descent('f1', lr).plotDescent()
    print('')
```

#### f2(x) for various learning rates

```python
for lr in [1.5, 1.0, 0.5, 0.1, 0.05, 0.01, 0.005, 0.001, 0.0005, 0.0001]:
    gradient_descent('f2', lr).plotDescent()
    print('')
```
